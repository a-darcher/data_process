```python
def clip_to_playback(vector, rec_on, rec_off):
    playback_spks = []
    for j, val in enumerate(vector):
        if val >= rec_on and val <= rec_off:
            playback_spks.append(val)
            
    return playback_spks


def get_all_spikes_mg_data(file_path, type_prefix, rec_on=None, rec_off=None):
    
    spikes = []
    
    trbl_key = [] #for troubleshooting
    
    for filename in os.listdir(file_path):
            if filename.startswith(type_prefix):
                #binary = np.load(os.path.join(path_binaries, str(pat_id), filename))    
                    if rec_on: 
                        times = sio.loadmat(os.path.join(file_path, filename)) 
                        times = times["cluster_class"]

                        ##### for trouble-shooting #####
                        num = filename.split("CSC")[1]
                        csc_file = int(num[:-4])
                        #csc = sio.loadmat(path + "CSC{}_spikes.mat".format(csc_file))
                        clus_info = sio.loadmat(file_path + "cluster_info.mat")
                        clus_key = clus_info['cluster_info'][0][csc_file - 1]
                        trbl_key.append([csc_file, clus_key])
                        #trbl_key.append(csc_file)
                        #separate out non-artifact spikes (currently not concerned with whether they are clustered into separate units)

                        labels = [row[0] for i,row in enumerate(times)]
                        set_ = set(labels)

                        unit_key = []

                        if len(set_) == 6:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 2:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 3:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 4:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 5:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)


                        if len(set_) == 5:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 2:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 3:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 4:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                        if len(set_) == 4:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 2:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 3:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)


                        if len(set_) == 3:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 2:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)

                        if len(set_) == 2:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = clip_to_playback(all_unit_spks, rec_on, rec_off)
                            spikes.append(all_unit_mv)
                        
                    if rec_on is None: 
                        times = sio.loadmat(os.path.join(file_path, filename)) 
                        times = times["cluster_class"]

                        ##### for trouble-shooting #####
                        num = filename.split("CSC")[1]
                        csc_file = int(num[:-4])
                        #csc = sio.loadmat(path + "CSC{}_spikes.mat".format(csc_file))
                        clus_info = sio.loadmat(file_path + "cluster_info.mat")
                        clus_key = clus_info['cluster_info'][0][csc_file - 1]
                        trbl_key.append([csc_file, clus_key])
                        #trbl_key.append(csc_file)
                        #separate out non-artifact spikes (currently not concerned with whether they are clustered into separate units)

                        labels = [row[0] for i,row in enumerate(times)]
                        set_ = set(labels)

                        unit_key = []

                        if len(set_) == 6:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 2:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 3:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 4:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 5:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)


                        if len(set_) == 5:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 2:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 3:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 4:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                        if len(set_) == 4:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 2:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 3:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)


                        if len(set_) == 3:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 2:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)

                        if len(set_) == 2:
                            all_unit_spks = []
                            for i, pair in enumerate(times):
                                if int(pair[0]) == 1:
                                    all_unit_spks.append(pair[1])
                            all_unit_mv = all_unit_spks
                            spikes.append(all_unit_mv)
    
    return spikes, trbl_key

## text related functions

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text[0]) ]
```

Plot binaries:

```python
def plot_raster(all_data, pat_id, save=None):

    fig = plt.figure(figsize=(80,20))
    ax1 = fig.add_subplot(121)
    ax1.eventplot(all_data, linewidths=0.1, linelengths=1.2)

    ratio = 0.3
    xleft, xright = ax1.get_xlim()
    ybottom, ytop = ax1.get_ylim()

    ax1.set_aspect(abs((xright-xleft)/(ybottom-ytop))*ratio)
    ax1.set_yticks([])
    #ax1.xaxis.set_tick_params(labelsize=30)
    ax1.tick_params(axis='x', labelsize=15 )

    plt.title('Spikes, Patient {}'.format(pat_id), size=30)
    plt.xlabel('Time (msec)', size=20)
    plt.ylabel('Units', size=20)

    ax1.spines['left'].set_visible(False)
    ax1.spines['right'].set_visible(False)
    ax1.spines['top'].set_visible(False)

    plt.tight_layout()

    if save:
        plt.savefig(save + "{}_raster.png".format(pat_id), bbox_inches="tight")
        #plt.savefig(save + "{}_raster.svg".format(pat_id), bbox_inches="tight")

    plt.show()
```
